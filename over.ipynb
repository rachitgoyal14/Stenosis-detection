```python
import cv2
import numpy as np
from skimage.morphology import skeletonize
import matplotlib.pyplot as plt

roi_path  = r"roi\14_002_5_0025_bmp.rf.debdb2e5cdec339506b1e5944ca03feb.jpg"
mask_path = r"masks\14_002_5_0025_bmp.rf.debdb2e5cdec339506b1e5944ca03feb_mask.png"

roi = cv2.imread(roi_path)
mask = cv2.imread(mask_path, 0)

# Clean mask again to be safe
binary = mask > 0
skeleton = skeletonize(binary)

# Overlay
overlay = roi.copy()
overlay[binary] = (0, 255, 0)     # lumen in green
overlay[skeleton] = (0, 0, 255)   # centerline in red

plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB))
plt.title("Lumen (Green) + Centerline (Red)")
plt.axis("off")
plt.show()
```

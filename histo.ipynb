```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("results/stenosis_results.csv")

plt.figure(figsize=(6,4))
plt.hist(df["stenosis_percent"], bins=10)
plt.xlabel("Stenosis %")
plt.ylabel("Number of images")
plt.title("Distribution of Stenosis Severity")
plt.grid(True)
plt.show()
```
